# 2. Weekly graph question

Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

_Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities_

There are numberous preattentive attributes that can help draw the viewer's attention to the first three months of data. We could chage the color of that portion of the graph to red, so it sticks out, especially compared the rest of the blue graph. Alternatively, we could make that section of the graph thicker and the rest of the line thinner or shade the background of that section of the graph in yellow to draw the viewer's attention. In addition, we could add a label to that portion of the graph to ensure our viewer pays attention to that section.

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt

categories = ['Engine power is less than expected', 'Tires make excessive noise while driving', 'Engine makes abnormal/excessive noise', 'Seat material concerns', 'Excessive wind noise', 'Hesitation or delay when shifting', 'Bluetooth system has poor sound quality', 'Steering system/wheel has too much play', 'Bluetooth system is difficult to use', 'Front seat audio/entertainment/navigation controls']
values = [12.9, 12.3, 11.6, 11.6, 11.0, 10.3, 10.0, 8.8, 8.6, 8.2]

categories.reverse()
values.reverse()

bars = plt.barh(categories, values, color='black')

for bar in bars:
    plt.text(bar.get_width() - bar.get_width()*0.1, bar.get_y() + bar.get_height()/2,
             str(bar.get_width()), va='center', ha='right', color='white')
            
plt.title("Top 10 design concerns")
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```
